```python
import os
import json
import requests
from pathlib import Path
from base64 import b64encode
from dotenv import load_dotenv
load_dotenv()
```

```python
url = "https://services.ceda.ac.uk/api/token/create/"

username = os.environ['CEDA_USERNAME']
password = os.environ['CEDA_PASSWORD']
token = b64encode(f"{username}:{password}".encode('utf-8')).decode("ascii")
headers = {
    "Authorization": f'Basic {token}',
}

response = requests.request("POST", url, headers=headers)

# If successful, this will return a JSON response containing the token
response_data = json.loads(response.text)
if response.status_code == 200:
    token = response_data["access_token"]
```

Daily:

```python
filename = f'{var}_hadukgrid_uk_1km_day_{year}{month:02}01-{year}{month:02}{ndays}.nc'
response = requests.request("GET", url + filename, headers=headers, stream=True)
with open('./test.nc', 'wb') as f:
    f.write(response.content)
    
for var in ['tasmin', 'tasmax']:
    print(var)
    for year in range(2005, 2023):
        print(year)
        for month in range(1,13):
            print(f'  {month:02}')
            #Â Handle leap years:
            if year % 4 == 0 and month == 2:
                ndays = 29
            else:
                ndays = int('_ 31 28 31 30 31 30 31 31 30 31 30 31'.split()[month])
            # Move to correct location on FTP server:
            url = f"https://dap.ceda.ac.uk/badc/ukmo-hadobs/data/insitu/MOHC/HadOBS/HadUK-Grid/v1.2.0.ceda/1km/{var}/day/v20230328/"
            # Define data source and destination:
            filename = f'{var}_hadukgrid_uk_1km_day_{year}{month:02}01-{year}{month:02}{ndays}.nc'
            dest = f'./haduk-data/daily/{filename}'
            response = requests.request("GET", url + filename, headers=headers, stream=True)
            if response.status_code == 200:
                with open(dest, 'wb') as f:
                    f.write(response.content)
            else:
                print(filename)
                print(response.status_code)
                print('---')
```
